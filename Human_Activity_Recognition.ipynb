# Human Activity Recognition

* Millions of People on Earth have a Smartphone available with them, and in addition to the cameras, smartphones also have the presence of many sensors in them. <b> These Sensors are extremely useful especially when we are playing Games on our Smartphones. </b>
* Two such Sensors on our Smartphones are Accelerometer and a Gyroscope : <b> *An Accelerometer measures the Acceleration whereas a Gyroscope measures the Angular Velocity.* </b>
* Now given that we have a Smartphone and we are able to collect the Accelerometer as well as the Gyroscope Data, the question is : <b> Can we somehow use this data to predict the activity that the person is performing? </b>
* The bunch of activities that a person is performing includes Walking, Walking Upstairs, Walking Downstairs, Sitting Down, Standing Up, Laying Down etc. :- A Total of 6 Activities.

<h2> Q. Why is this Useful in the First Place? </h2>

* There are very popular Smart watches like Fitbit, Apple Watch etc., which also internally contain an accelerometer and a gyroscope. This data is actually used to determine the activity that you are carrying out and to track your activities throughout the day.
* The advantage is that after all of this data is collected by the Smart watches, the devices predict the amount of calories that have been burnt by the user throughout the course of the day. Similarly these devices can also measure the number of hours that the user has slept, because if you are laying down, it is most likely that you are sleeping. 
* <b> *However, it is to be noted that most of the Modern Smartphones have an additional sensor called a Heart-Rate Sensor, which is used in addition to accelerometers and gyroscopes to determine if you are actually sleeping or not.* </b> *The earlier versions of these devices did not have Heart Rate Monitors because these sensors were expensive at that period of time. Also, we cannot have Heart Rate Monitors on our Smartphones.*

<h2> Mathematical Understanding of the Problem :</h2>

* Most of these Accelerometers that we have are Tri-Axial Accelerometers which means that the acceleration is measured across the X-axis, Y-axis and Z-axis. Similarly the Gyroscopes measure the Angular Velocity across the X-axis, Y-axis and Z-axis. 
* In the Accelerometer, we have the data with the X-axis acceleration measured over time. Similarly we have the acceleration on Y-axis measured over time and acceleration on Z-axis measured over time. Similarly for the Gyroscope, we have Angular Velocity measurements across X-axis, Y-axis and Z-axis. <b> We basically have a total of 6 Time-Series Data. </b>
* This 6 Time-Series Data becomes our Input, using which we want to predict our Output. The output that we want is the Activity : one of the 6 classes. <b> At the outset, it obviously looks like a Multi-class Classification Problem. </b>
* Some of the Metrics that we use in case of a Multi-class Classification Problem is the Overall Accuracy, as well as the Confusion Matrices, and the Multi-class log loss.
* The Confusion Matrices are important because they give information about the type of confusions that are taking place. Eg : Our Classification Model could get confused between 'Sitting' and 'Standing'.

## How was this Data Processed?

### Feature Names

* Assume that you have a data that depicts the acceleration along the x-axis being plotted against time. <b> *These sensor signals are preprocessed by applying noise filters and then sampled in fixed-width windows(sliding windows) of 2.56 seconds each with 50% overlap. ie., each window has 128 readings. This means that each of the vectors that we obtain is of dimensionality equal to 128.* </b>

* From Each window, a feature vector was obtained by calculating variables from the time and frequency domain. The vector thus obtained is of a fixed length. Now we can sample this and obtain values at a fixed interval.
* If we were to solve this problem in the real world, we would have worked with experts in signal processing who would have helped us with these filtering operations.
* Now we have already obtained our first vector. <b> *The second vector is not obtained by directly going to the second window.* </b> Now what we do is that we take the second window that overlaps this first window by exactly 50%.
* This means that instead of taking a window from time (2.56s to 2 * 2.56s), we are taking a window from (1.28s to 1.28 + 2.56). <b> This entire technique is called Overlappped Sampling, wherein the second window is overlapping the first window by 50%. </b> Now for this window obtained, we apply some signal processing filters, carry out some sampling, and convert it into a numerical vector of a fixed size.

***This entire technique that we discussed above is not carried out not just for Acc(x) ie. acceleration along the x-axis, but for all the time series datasets that we have.***

* In the original dataset we have accelerations along the x, y and z axes and the angular velocities along the x, y and z axes. *All of these are Time Series Datasets. We have a total of 6 time series datasets.*
* The people who created this dataset are domain experts, who know in depth about Human Activity Recognition, about Accelerometers, Gyroscopes etc. Also, they have a deep understanding of how the time series datasets look like. 
* Therefore, in addition to providing us with the 6 Time Series Datasets, these domain experts have designed/engineered features for us. <b> Some of these engineered features could be very complex because they involve a lot of Signal Processing stuff. </b>
* These features are as shown below. Eg : 'tBodyAccJerk-XYZ_'. The Jerk over here refers to the rate of change of acceleration, whereas the definition of acceleration is rate of change of velocity.

***For all of these manually engineered features, all of the features that are 'starting with t' are Time Based Features and all the features that are 'starting with f' are after applying the FFT.***


> In our dataset, each datapoint represents a window with different readings 

* The accelertion signal was saperated into Body and Gravity acceleration signals(___tBodyAcc-XYZ___ and ___tGravityAcc-XYZ___) using some low pass filter with corner frequecy of 0.3Hz.

4. After that, the body linear acceleration and angular velocity were derived in time to obtian _jerk signals_ (___tBodyAccJerk-XYZ___ and ___tBodyGyroJerk-XYZ___). 

5. The magnitude of these 3-dimensional signals were calculated using the Euclidian norm. This magnitudes are represented as features with names like _tBodyAccMag_, _tGravityAccMag_, _tBodyAccJerkMag_, _tBodyGyroMag_ and _tBodyGyroJerkMag_.

6. Finally, We've got frequency domain signals from some of the available signals by applying a FFT (Fast Fourier Transform). These signals obtained were labeled with ___prefix 'f'___ just like original signals with ___prefix 't'___. These signals are labeled as ___fBodyAcc-XYZ___, ___fBodyGyroMag___ etc.,.

8. Our Raw Data is encoded as a 128 dimensional vector for every window that we have, and on top of this vector we have computed a lot of features. We can esitmate some set of variables from the above signals. ie., <b> We will estimate the following properties on each and every signal that we recoreded so far. </b>

	+ ___mean()___: Mean value
	+ ___std()___: Standard deviation
	+ ___mad()___: Median absolute deviation 
	+ ___max()___: Largest value in array
	+ ___min()___: Smallest value in array
	+ ___sma()___: Signal magnitude area
	+ ___energy()___: Energy measure. Sum of the squares divided by the number of values. 
	+ ___iqr()___: Interquartile range 
	+ ___entropy()___: Signal entropy
	+ ___arCoeff()___: Autorregresion coefficients with Burg order equal to 4
	+ ___correlation()___: correlation coefficient between two signals
	+ ___maxInds()___: index of the frequency component with largest magnitude
	+ ___meanFreq()___: Weighted average of the frequency components to obtain a mean frequency
	+ ___skewness()___: skewness of the frequency domain signal 
	+ ___kurtosis()___: kurtosis of the frequency domain signal 
	+ ___bandsEnergy()___: Energy of a frequency interval within the 64 bins of the FFT of each window.
	+ ___angle()___: Angle between to vectors.

# 4. Applying T-SNE on the Data

* Note that the entire data that we have is in 561 dimensions, and in order to visualize data in such a case, we need to carry out dimensionality reduction. This is what we try and achieve with the help of T-SNE.
* Since we have 561 Dimensional Representation of Expert Engineered Features, and as we have seen above in the Univariate Analysis, that some single features themselves are able to partially separate the class labels, we will try and understand what we can achieve by incorporating all the 561 dimensions, the tool to achieve this being T-SNE.
* We are plotting into a 2-Dimensional space.

```python
import numpy as np
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# performs t-sne with different perplexity values and their repective plots..

def perform_tsne(X_data, y_data, perplexities, n_iter=1000, img_name_prefix='t-sne'):
        
    for index,perplexity in enumerate(perplexities):
        # perform t-sne
        print('\nperforming tsne with perplexity {} and with {} iterations at max'.format(perplexity, n_iter))
        X_reduced = TSNE(verbose=2, perplexity=perplexity).fit_transform(X_data)
        print('Done..')
        
        # prepare the data for seaborn         
        print('Creating plot for this t-sne visualization..')
        df = pd.DataFrame({'x':X_reduced[:,0], 'y':X_reduced[:,1] ,'label':y_data})
        
        # draw the plot in appropriate place in the grid
        sns.lmplot(data=df, x='x', y='y', hue='label', fit_reg=False, size=8,\
                   palette="Set1",markers=['^','v','s','o', '1','2'])
        plt.title("perplexity : {} and max_iter : {}".format(perplexity, n_iter))
        img_name = img_name_prefix + '_perp_{}_iter_{}.png'.format(perplexity, n_iter)
        print('saving this plot as image in present working directory...')
        plt.savefig(img_name)
        plt.show()
        print('Done')
```

```python
X_pre_tsne = train_final.drop(['subject', 'Activity','ActivityName'], axis=1)
y_pre_tsne = train_final['ActivityName']
perform_tsne(X_data = X_pre_tsne,y_data=y_pre_tsne, perplexities =[2,5,10,20,50])
```

__Observations__:

* As we can see above, the best separated plot that we obtain is for Perplexity=  2 and when we have run the same for 1000 iterations.
* One of the first observations that we can make from this plot is that most of the points are well clustered together. <b> There is not too much of overlap among our datapoints except in the case of 'Sitting' & 'Standing'. </b> *There is a region where the blue and red points do not overlap, whereas there is also a considerable region where the overlap is present.*

```python
import itertools
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
plt.rcParams["font.family"] = 'DejaVu Sans'

#function to plot the confusion matrix
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=90)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
```

### 5.3 Generic function to run any model specified

```python
from datetime import datetime
def perform_model(model, X_train, y_train, X_test, y_test, class_labels, cm_normalize=True, \
                 print_cm=True, cm_cmap=plt.cm.Greens):
    
    
    # to store results at various phases
    results = dict()
    
    # time at which model starts training 
    train_start_time = datetime.now()
    print('Training the model..')
    model.fit(X_train, y_train)
    print('Done \n \n')
    train_end_time = datetime.now()
    results['training_time'] =  train_end_time - train_start_time
    print('Training Time (HH:MM:SS.ms) - {}\n\n'.format(results['training_time']))
    
    
    # predict test data
    print('Predicting Test Data')
    test_start_time = datetime.now()
    y_pred = model.predict(X_test)
    test_end_time = datetime.now()
    print('Done \n \n')
    results['testing_time'] = test_end_time - test_start_time
    print('Testing Time (HH:MM:SS:ms) - {}\n\n'.format(results['testing_time']))
    results['predicted'] = y_pred
   

    # calculate overall accuracty of the model
    accuracy = metrics.accuracy_score(y_true=y_test, y_pred=y_pred)
    # store accuracy in results
    results['accuracy'] = accuracy
    print('---------------------')
    print('|      Accuracy      |')
    print('---------------------')
    print('\n    {}\n\n'.format(accuracy))
    
    
    # confusion matrix
    cm = metrics.confusion_matrix(y_test, y_pred)
    results['confusion_matrix'] = cm
    if print_cm: 
        print('--------------------')
        print('| Confusion Matrix |')
        print('--------------------')
        print('\n {}'.format(cm))
        
    # plot confusin matrix
    plt.figure(figsize=(8,8))
    plt.grid(b=False)
    plot_confusion_matrix(cm, classes=class_labels, normalize=True, title='Normalized confusion matrix', cmap = cm_cmap)
    plt.show()
    
    # get classification report
    print('-------------------------')
    print('| Classifiction Report |')
    print('-------------------------')
    classification_report = metrics.classification_report(y_test, y_pred)
    # store report in results
    results['classification_report'] = classification_report
    print(classification_report)
    
    # add the trained  model to the results
    results['model'] = model
    
    return results
    
```

### 5.4 Method to print the GridSearch Attributes

# 7. Applying Deep Learning Models on the Raw Data

* Here, instead of 6 Time Series Datasets, we actually have a total of 9 Time Series Datasets, where 6 of these Time Series Datasets are for the Accelerometer Readings whereas 3 of them are for the Gyroscope Readings : This is the Actual/Raw Time Series Data.
* All of these Time Series Datasets start at Time=0, and we break each time series into overlapping windows of 2.56 seconds (overlap of 50%). Each window corresponded to the 561 engineered features in the previous example.
* We take each window, perform some signal processing on top of that and encode the same as a 128-dimensional vector.
* Thus, at any Time t, we have 9 readings ie. we have a vector of 9 values. We are not designing any features over here and are just taking the raw windows :
        Each window has 128 time steps. At every Time step, we have a total of 9 recordings, one corresponding to each of the Time Series.

```python
from tensorflow.python.keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
from datetime import datetime
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dropout, Dense, Flatten, BatchNormalization
```

```python
# Activities are the class labels
# It is a 6 class classification
ACTIVITIES = {
    1: 'WALKING',
    2: 'WALKING_UPSTAIRS',
    3: 'WALKING_DOWNSTAIRS',
    4: 'SITTING',
    5: 'STANDING',
    6: 'LAYING',
}

# Utility function to print the confusion matrix
def confusion_matrix(Y_true, Y_pred):
    Y_true = pd.Series([ACTIVITIES[y] for y in np.argmax(Y_true, axis=1)])
    Y_pred = pd.Series([ACTIVITIES[y] for y in np.argmax(Y_pred, axis=1)])

    return pd.crosstab(Y_true, Y_pred, rownames=['True'], colnames=['Pred'])
```

```python
##gives train and validation data 
def data():
    """
    Obtain the dataset from multiple files.
    Returns: X_train, X_test, y_train, y_test
    """
    # Data directory
    DATADIR = 'UCI_HAR_Dataset'
    SIGNALS = [
        "body_acc_x",
        "body_acc_y",
        "body_acc_z",
        "body_gyro_x",
        "body_gyro_y",
        "body_gyro_z",
        "total_acc_x",
        "total_acc_y",
        "total_acc_z"
        ]
    # Utility function to read the data from csv file
    def _read_csv(filename):
        return pd.read_csv(filename, delim_whitespace=True, header=None)

    # Utility function to load the load
    def load_signals(subset):
        signals_data = []

        for signal in SIGNALS:
            filename = f'HAR/UCI_HAR_Dataset/{subset}/Inertial Signals/{signal}_{subset}.txt'
            signals_data.append( _read_csv(filename).to_numpy()) 

        # Transpose is used to change the dimensionality of the output,
        # aggregating the signals by combination of sample/timestep.
        # Resultant shape is (7352 train/2947 test samples, 128 timesteps, 9 signals)
        return np.transpose(signals_data, (1, 2, 0))

    def load_y(subset):
        """
        The objective that we are trying to predict is a integer, from 1 to 6,
        that represents a human activity. We return a binary representation of 
        every sample objective as a 6 bits vector using One Hot Encoding
        (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)
        """
        filename = f'HAR/UCI_HAR_Dataset/{subset}/y_{subset}.txt'
        y = _read_csv(filename)[0]
        return pd.get_dummies(y).to_numpy()
    
    X_train, X_test = load_signals('train'), load_signals('test')
    Y_train, Y_test = load_y('train'), load_y('test')

    return X_train, Y_train, X_test,  Y_test
```

```python
# Utility function to count the number of classes
def _count_classes(y):
    return len(set([tuple(category) for category in y]))

def plt_dynamic(x, vy, ty, ax, colors=['b']):
    ax.plot(x, vy, 'b', label="Test Accuracy")
    ax.plot(x, ty, 'r', label="Train Accuracy")
    plt.legend()
    plt.grid()
    fig.canvas.draw()
```

```python
#Function to plot the confusion matrix : Both Normalized and Non-Normalized Confusion Matrices

def plot_lstm_confusion_matrix(model,normalize=False,):
    
    y_true = pd.Series([ACTIVITIES[y+1] for y in np.argmax(Y_test, axis=1)])
    y_pred = model.predict(X_test)
    y_pred_final = pd.Series([ACTIVITIES[y+1] for y in np.argmax(y_pred, axis=1)])
    
    plt.figure(figsize=(8,8))
    plt.grid(b=False)
    
    if normalize:
        conf_matrix = pd.crosstab(y_true, y_pred_final, rownames=['True label'], colnames=['Predicted label'])
        conf_matrix = conf_matrix.astype('float') / conf_matrix.sum(axis=1)[:, np.newaxis]
        sns.heatmap(conf_matrix, annot=True, cmap="Greens", fmt=".3f", xticklabels=labels, yticklabels=labels)
        plt.title('NORMALIZED CONFUSION MATRIX')
    else:
        conf_matrix = pd.crosstab(y_true, y_pred_final, rownames=['True label'], colnames=['Predicted label'])
        sns.heatmap(conf_matrix, annot=True, cmap="Blues", fmt=".3f", xticklabels=labels, yticklabels=labels)
        plt.title('NON-NORMALIZED CONFUSION MATRIX')
    
    plt.show()
```

# 8. Divide and Conquer CNN 

Refer :- https://www.researchgate.net/publication/324224939_Divide_and_Conquer-Based_1D_CNN_Human_Activity_Recognition_Using_Test_Data_Sharpening

In the 'Divide and Conquer CNN' based approach we first build a CNN Model that predicts whether the activity is static or dynamic with a very high accuracy. This is followed by building models to separately predict the static and dynamic activity in consideration.

## 8.1 Classification into 2 Classes : Static and Dynamic Activities

#### 8.1.1 Function to obtain Train and Test Data after Classifying Labels into Static and Dynamic Activities

```python
def static_dynamic_classification():
    """
    Obtain the dataset from multiple files.
    Returns: X_train, X_test, y_train, y_test
    """
    # Data directory
    DATADIR = 'UCI_HAR_Dataset'
    SIGNALS = [
        "body_acc_x",
        "body_acc_y",
        "body_acc_z",
        "body_gyro_x",
        "body_gyro_y",
        "body_gyro_z",
        "total_acc_x",
        "total_acc_y",
        "total_acc_z"
        ]
    # Utility function to read the data from csv file
    def _read_csv(filename):
        return pd.read_csv(filename, delim_whitespace=True, header=None)

    # Utility function to load the load
    def load_signals(subset):
        signals_data = []

        for signal in SIGNALS:
            filename = f'HAR/UCI_HAR_Dataset/{subset}/Inertial Signals/{signal}_{subset}.txt'
            signals_data.append( _read_csv(filename).to_numpy()) 

        # Transpose is used to change the dimensionality of the output,
        # aggregating the signals by combination of sample/timestep.
        # Resultant shape is (7352 train/2947 test samples, 128 timesteps, 9 signals)
        return np.transpose(signals_data, (1, 2, 0))
    
    def load_y(subset):
        """
        The objective that we are trying to predict is a integer, from 1 to 6,
        that represents a human activity. We return a binary representation of 
        every sample objective as a 6 bits vector using One Hot Encoding
        (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)
        """
        filename = f'HAR/UCI_HAR_Dataset/{subset}/y_{subset}.txt'
        y = _read_csv(filename)[0]
        y = (y>3).astype(int)
        
        return pd.get_dummies(y).to_numpy()
    
    X_train_2, X_test_2 = load_signals('train'), load_signals('test')
    Y_train_2, Y_test_2 = load_y('train'), load_y('test')

    return X_train_2, Y_train_2, X_test_2,  Y_test_2
```

#### 8.3.1 Function to obtain Train and Test Data for Dynamic Activity Classification

```python
def dynamic_classification():
    """
    Obtain the dataset from multiple files.
    Returns: X_train, X_test, y_train, y_test
    """
    # Data directory
    DATADIR = 'UCI_HAR_Dataset'
    SIGNALS = [
        "body_acc_x",
        "body_acc_y",
        "body_acc_z",
        "body_gyro_x",
        "body_gyro_y",
        "body_gyro_z",
        "total_acc_x",
        "total_acc_y",
        "total_acc_z"
        ]
    # Utility function to read the data from csv file
    def _read_csv(filename):
        return pd.read_csv(filename, delim_whitespace=True, header=None)

    # Utility function to load the load
    def load_signals(subset):
        signals_data = []

        for signal in SIGNALS:
            filename = f'HAR/UCI_HAR_Dataset/{subset}/Inertial Signals/{signal}_{subset}.txt'
            signals_data.append( _read_csv(filename).to_numpy()) 

        # Transpose is used to change the dimensionality of the output,
        # aggregating the signals by combination of sample/timestep.
        # Resultant shape is (7352 train/2947 test samples, 128 timesteps, 9 signals)
        return np.transpose(signals_data, (1, 2, 0))
    
    def load_y(subset):
        """
        The objective that we are trying to predict is a integer, from 1 to 6,
        that represents a human activity. We return a binary representation of 
        every sample objective as a 6 bits vector using One Hot Encoding
        (https://pandas.pydata.org/pandas-docs/stable/generated/pandas.get_dummies.html)
        """
        filename = f'HAR/UCI_HAR_Dataset/{subset}/y_{subset}.txt'
        y = _read_csv(filename)[0]
        y_filter = (y<=3)
        y_final = y[y_filter]
        
        return pd.get_dummies(y_final).to_numpy(), y_filter
    
   
    Y_train_dy, Y_train_filter = load_y('train')
    Y_test_dy, Y_test_filter = load_y('test')
    
    X_train_dy, X_test_dy = load_signals('train'), load_signals('test')
    X_train_dy = X_train_dy[Y_train_filter]
    X_test_dy = X_test_dy[Y_test_filter]

    return X_train_dy, Y_train_dy, X_test_dy,  Y_test_dy
```

```python
X_train_dy, Y_train_dy, X_test_dy, Y_test_dy = dynamic_classification()

print('Shape of the Training Data:')
print(" "*100)
print("Shape of X: " + str(X_train_dy.shape))
print("Shape of Y: " + str(Y_train_dy.shape))
print(" "*100)

print("-"*100)
print(" "*100)

print('Shape of the Test Data:')
print(" "*100)
print("Shape of X: " + str(X_test_dy.shape))
print("Shape of Y: " + str(Y_test_dy.shape))
```

### 8.3.2 Model to Classify Dynamic Activities

<h4> 8.3.2.1 Constructing the Neural Network </h4>

```python
model_dynamic = Sequential()
model_dynamic.add(Conv1D(64, kernel_size=3,
                 activation='relu', kernel_initializer='he_uniform',
                 input_shape=(128,9), padding='same'))
model_dynamic.add(Conv1D(64, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(MaxPooling1D(pool_size=2))
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Conv1D(128, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(Conv1D(128, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(MaxPooling1D(pool_size=2))
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Conv1D(256, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(Conv1D(256, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(Conv1D(256, kernel_size = 3, activation='relu',kernel_initializer='he_uniform', padding='same'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(MaxPooling1D(pool_size=2))
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Flatten())

model_dynamic.add(Dense(512, activation='relu',kernel_initializer='he_uniform'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Dense(256, activation='relu',kernel_initializer='he_uniform'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Dense(128, activation='relu',kernel_initializer='he_uniform'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Dense(64, activation='relu',kernel_initializer='he_uniform'))
model_dynamic.add(BatchNormalization())
model_dynamic.add(Dropout(0.5))

model_dynamic.add(Dense(3, activation='softmax',kernel_initializer='he_uniform'))
model_dynamic.summary()
```

<h4> 8.3.2.2 Running the Neural Network on Train & Test Datasets for 30 Epochs </h4>

```python
#Refer :- https://github.com/tensorflow/tensorflow/issues/27023

import tensorflow.python.util.deprecation as deprecation
deprecation._PRINT_DEPRECATION_WARNINGS = False

#Compiling the Model
model_dynamic.compile(optimizer='rmsprop', loss='categorical_crossentropy', metrics=['accuracy'])

#Training the Model
M_dynamic_history = model_dynamic.fit(X_train_dy, Y_train_dy, batch_size=256, epochs=30, verbose=1, \
                       validation_data=(X_test_dy, Y_test_dy)) 

#Saving the model for later use
model_dynamic.save('HAR/UCI_HAR_Dataset/dynamic_classification.h5')
```

```python
# Refer :- https://machinelearningmastery.com/deep-learning-models-for-human-activity-recognition/
# Refer :- http://www.mdpi.com/1424-8220/18/4/1055 

type_predict = type_classifier.predict(X_test_2) #returns the class probabilities for individual classes
predicted_classes = np.argmax(type_predict, axis=1) #conversion of probabilities to actual class
actual_classes = Y_test_2.flatten() #flatten 2D array to 1D

static_classes = X_test_2[predicted_classes==1] #obtaining X features of Test Data for static
dynamic_classes = X_test_2[predicted_classes==0] #obtaining X features of Test Data for dynamic
```

```python
static_prediction = static_model.predict(static_classes) #obtaining probabilities for the static classes
static_class_prediction = np.argmax(static_prediction,axis=1) #obtaining actual class labels
static_class_prediction = static_class_prediction + 4  
#class labels obtained are from 0 to 3. Need to add 4 to obtain class labels from 4 to 6 (static class labels).

dynamic_prediction = dynamic_model.predict(dynamic_classes) #obtaining probabilities for the dynamic classes
dynamic_class_prediction = np.argmax(dynamic_prediction, axis=1) #obtaining actual class labels
dynamic_class_prediction = dynamic_class_prediction + 1 
#class labels obtained are from 0 to 2. Need to add 1 to obtain class labels from 1 to 3 (dynamic class labels).

#obtaining class labels in the order of their occurrence
i,j = 0,0 
final_predicted_classes = []
for label in predicted_classes:
    if label == 1:
        final_predicted_classes.append(static_class_prediction[i])
        i = i + 1
    elif label== 0:
        final_predicted_classes.append(dynamic_class_prediction[j])
        j = j + 1
```

```python
final_predicted_classes = pd.get_dummies(final_predicted_classes).to_numpy()  
#one hot encoding of final predicted classes 

from sklearn.metrics import accuracy_score
print('Accuracy of test data',accuracy_score(Y_test,final_predicted_classes))  #final accuracy score
```

The Human Activity Recognition is a very popular application of modern times, especially with the advent of Smartphones as well as Smartwatches that have become extremely prevalent. The data is recorded with the help of Accelerometers (that measure acceleration) as well as Gyroscopes (that measure angular velocity). We have a total of 6 time series datasets. Ours is a 6-class classification task for the 6 activities that we have. Since this is a Multiclass Classification Task, the metrics that we use are log-loss, accuracy and confusion matrices, with accuracy being the primary metric that we focus on.

In the dataset we carry out Overlapped Sampling (50% overlap), at the end of which we obtain 128 dimensional vectors, and we have a total of 561 Expert Engineered Features on which we employed various Machine Learning Models, and as we can see in the summary below, our linear models have performed much better than the tree based methodologies. However, at the same time, we see from the Confusion Matrix that there is a huge confusion between the 'Sitting' and 'Standing' classes.

At any Time t, we have 9 readings ie. we have a vector of 9 values. Each window has 128 time steps. At every Time step, we have a total of 9 recordings, one corresponding to each of the Time Series.
